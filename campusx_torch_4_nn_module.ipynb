```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32' ], inplace = True)
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0], test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
encoder  = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

### Model Creation 

```python
class MySimpleNN(nn.Module): # Use OOPs concept --> build a class 
    
    def __init__(self,num_features ): # Build a constructor 

        super().__init__() # inheritence 
        self.linear= nn.Linear(num_features, 1)
        # self.relu = nn.ReLU()
        # self.linear2 = nn.Linear(3,1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, features):

        out = self.linear(features)
        # out = self.relu(out)
        # out = self.linear2(out)
        out = self.sigmoid(out)

        return out

    def loss_function(self, y_pred, y):
        # Clamp predictions to avoid log(0)
        epsilon = 1e-7
        y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

        # Calculate loss
        loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
        return loss
```

```python
learning_rate  = 0.1
epochs  = 25
```

### training pipeline 

```python
X_train_tensor = X_train_tensor.float()
y_train_tensor = y_train_tensor.float()  # Or `.long()` if these are class labels
```

```python
#create model 
model = MySimpleNN(X_train_tensor.shape[1])

# Create a dataset

#features  = torch.rand(X_train_tensor.shape[1])


## Define Loop

for epoch in range(epochs):
    

## Forward Pass 
    y_pred = model(X_train_tensor)
    #print(y_pred)

## Loss Calculate 
    loss = model.loss_function(y_pred, y_train_tensor)
    #print(f"Epoch: {epoch + 1}, Loss: {loss.item()}")


## Backward Pass 
    loss.backward()

## Parameters Update
    with torch.no_grad(): # For not tracking the gradient 
        model.linear.weight -= learning_rate * model.linear.weight.grad
        model.linear.bias -= learning_rate * model.linear.bias.grad

## Zero gradients 
    model.linear.weight.grad.zero_()
    model.linear.bias.grad.zero_()

## Print Loss in each epoch
    print(f"Epoch: {epoch + 1}, Loss: {loss.item()}")

    
```

### Evaluation

```python
# Ensure data types match
X_test_tensor = X_test_tensor.float()
y_test_tensor = y_test_tensor.float()

# Evaluation with no gradient tracking
with torch.no_grad():
    y_pred = model.forward(X_test_tensor)
    y_pred = (y_pred > 0.9).float()
    accuracy = (y_pred == y_test_tensor).float().mean()

print(f'Accuracy: {accuracy.item():.4f}')
```

```python
# Create model class using nn.module 

import torch 
import torch.nn as nn

class Model(nn.Module):

    def __init__(self, num_features):

        super().__init__() # inheritence 
        self.linear1 = nn.Linear(num_features, 3)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(3,1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, features):

        out = self.linear1(features)
        out = self.relu(out)
        out = self.linear2(out)
        out = self.sigmoid(out)

        return out
```

### Using Sequential container (to stack layers in order)

```python
# Create model class using nn.module 

import torch 
import torch.nn as nn

class Model(nn.Module):

    def __init__(self, num_features):

        super().__init__() 
        self.network = nn.Sequential(
            nn.Linear(num_features, 3),
            nn.ReLU(),
            nn.Linear(3,1),
            nn.Sigmoid()
        )

    def forward(self, features):

        out = self.network(features)

        return out
```

```python
# Create a dataset

features  = torch.rand(10, 5)

# create a model 

model = Model(features.shape[1])

# call model for forward pass 

#model.forward(features) # No need to write like this 

model(features)
```
